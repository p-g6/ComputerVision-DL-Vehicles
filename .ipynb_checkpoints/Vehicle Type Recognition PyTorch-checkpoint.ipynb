```python
# imports
import torch
from torch import nn
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from sklearn.model_selection import train_test_split as tts

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(torch.__version__)
print(device)
```

```python
from pathlib import Path
data_path = Path("data/")
image_path = data_path / "VehicleTypeRecog"

# data from: https://www.kaggle.com/datasets/kaggleashwin/vehicle-type-recognition
```

```python
train_dataloader = DataLoader(dataset=train_data,
                              batch_size=32,
                              num_workers=os.cpu_count(),
                              shuffle=True)
test_dataloader = DataLoader(dataset=test_data,
                             batch_size=32,
                             num_workers=os.cpu_count(),
                             shuffle=False)
```

```python
img, label = next(iter(train_dataloader))
img.shape, label.shape
```

```python
class TinyVGG(nn.Module):
    def __init__(self, inputs, hiddens, outputs):
        super().__init__()
        self.layer_stack = nn.Sequential(
            nn.Conv2d(inputs, hiddens, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(hiddens, hiddens, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(hiddens, hiddens, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(hiddens, hiddens, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Flatten(),
            nn.Linear(hiddens*64*64, outputs)
        )
        
    def forward(self, x):
        return self.layer_stack(x)

torch.manual_seed(42)
m0 = TinyVGG(3, 10, len(class_names)).to(device)
# next(m0.parameters()).is_cuda to verify on gpu
m0 
```

```python
from torchinfo import summary
summary(m0, input_size=[32,3,256,256])
```

```python
from torchmetrics import Accuracy

accfn = Accuracy(task='multiclass', num_classes=len(class_names)).to(device)

def train_step(model, dataloader, lossfn, optimizer, accfn, device=device):
    model.train()
    trainloss, trainacc = 0, 0
    model.to(device)
    for batch, (X,y) in enumerate(dataloader):
        X, y = X.to(device), y.to(device)
        ypred = model(X)
        
        loss = lossfn(ypred, y)
        trainloss += loss
        trainacc += accfn(y, ypred.argmax(dim=1))
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    trainloss /= len(dataloader)
    trainacc /= len(dataloader)
    
    return trainloss, trainacc
```

```python
def test_step(model, dataloader, lossfn, accfn, device=device):
    model.to(device)
    model.eval()
    testloss, testacc = 0, 0
    with torch.inference_mode():
        for batch, (X,y) in enumerate(dataloader):
            X, y = X.to(device), y.to(device)
            testpred = model(X)
            testloss += lossfn(testpred, y)
            testacc += accfn(y, testpred.argmax(dim=1))
        
        testloss /= len(dataloader)
        testacc /= len(dataloader)
        
    return testloss, testacc
```

```python
from tqdm.auto import tqdm

def train(model, train_dataloader, test_dataloader, optimizer, lossfn, accfn, epochs):
    results = {"train_loss":[],
               'train_acc':[],
               'test_loss':[],
               'test_acc':[]}
    for epoch in tqdm(range(epochs)):
        train_loss, train_acc = train_step(model, train_dataloader, lossfn, optimizer, accfn)
        test_loss, test_acc = test_step(model, test_dataloader, lossfn, accfn)
        
        print(
            f"Epoch: {epoch+1} | "
            f"train_loss: {train_loss:.4f} | "
            f"train_acc: {train_acc:.4f} | "
            f"test_loss: {test_loss:.4f} | "
            f"test_acc: {test_acc:.4f}")
        
        results['train_loss'].append(train_loss)
        results['train_acc'].append(train_acc)
        results['test_loss'].append(test_loss)
        results['test_acc'].append(test_acc)
        
    return results
```

```python
torch.manual_seed(42)
torch.cuda.manual_seed(42)

lossfn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(m0.parameters(), lr=0.01)

from timeit import default_timer as timer
start = timer()

m0_results = train(m0, train_dataloader, test_dataloader, optimizer, lossfn, accfn, 10)
end = timer()

print(f"Total training time: {end-start:.3f} seconds")
```

```python
# trying VGG11 architecture

class VGG11(nn.Module):
    def __init__(self, in_channels, outputs):
        super().__init__()
        self.in_channels = in_channels
        # convolutional layers 
        self.conv_layers = nn.Sequential(
            nn.Conv2d(self.in_channels, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        # fully connected linear layers
        self.linear_layers = nn.Sequential(
            nn.Linear(in_features=512*8*8, out_features=4096),
            nn.ReLU(),
#             nn.Dropout2d(0.5),
            nn.Linear(in_features=4096, out_features=4096),
            nn.ReLU(),
#             nn.Dropout2d(0.5),
            nn.Linear(in_features=4096, out_features=outputs)
        )
    def forward(self, x):
        x = self.conv_layers(x)
        # flatten to prepare for the fully connected layers
        x = x.view(x.size(0), -1)
        x = self.linear_layers(x)
        return x
    
m1 = VGG11(3, 4).to(device)
m1
```

```python
# train and test of new model
torch.manual_seed(42)
torch.cuda.manual_seed(42)

lossfn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(m1.parameters(), lr=0.001)

from timeit import default_timer as timer
start = timer()

m0_results = train(m1, train_dataloader, test_dataloader, optimizer, lossfn, accfn, 10)
end = timer()

print(f"Total training time: {end-start:.3f} seconds")
```

```python
# trying efficientnet_b0 transfer learning
import torchvision

weights = torchvision.models.EfficientNet_B0_Weights.DEFAULT
m3 = torchvision.models.efficientnet_b0(weights=weights).to(device)
m3
```

```python
for param in m3.features.parameters():
    param.requires_grad = False
```

```python
torch.manual_seed(42)
torch.cuda.manual_seed(42)

output_shape = len(class_names)
m3.classifier = torch.nn.Sequential(
    nn.Dropout(p=0.2, inplace=True),
    nn.Linear(in_features=1280, out_features=output_shape, bias=True)
).to(device)
```

```python
lossfn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(m3.parameters(), lr=0.001)

from timeit import default_timer as timer 
start_time = timer()

# Setup training and save the results
results = train(model=m3,
                       train_dataloader=train_dataloader,
                       test_dataloader=test_dataloader,
                       optimizer=optimizer,
                       lossfn=lossfn,
                       accfn=accfn,
                       epochs=20)

# End the timer and print out how long it took
end_time = timer()
print(f"[INFO] Total training time: {end_time-start_time:.3f} seconds")
```
